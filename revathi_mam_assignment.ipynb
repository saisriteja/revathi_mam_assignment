#Plot N sample values of the speech file (using appropriateMATLAB plot commands),starting at sample fstart and ending with sample fstart+N-1. 

```
#function for plotting n samples
def plot_samples(start_,N,data):

  if start_+N > len(data):
    return 'not possible to draw because the length of the N is larger than the number of samples'

  plt.figure(figsize=(18,5))
  plt.plot(data[start_:start_+N-1])
  plt.show()
```

```
plot_samples(25000,10000,x)
```

#Plot M samples ofspeech per line, with up to four lines of plotting per page. The plotting parametersfstart, M, and N should all be specified at run time (i.e., they should be read into theprogram).

```
def plot_samples_4(fstart,M,data,sr = 22050:
  fig,a =  plt.subplots(4,1,figsize=(10, 10),squeeze=False,sharex=False)

  p = fstart 
  count = M

  for i in range(4):
    a[i][0].plot(data[p:p+count])
    a[i][0].set_title(f'data is in between samples {p}-{p+count} and time is in between {round(p/sr,2)}-{round((p+count)/sr,2)}')

    p = p+count

  fig.tight_layout()
  plt.show()



M = 2000
plot_samples_4(35000,M,x)
```
